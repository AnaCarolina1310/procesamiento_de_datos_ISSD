* Dicho texto puede ser descargado de Internet (webscrapping) o extraído del dataset de alguna librería.

* Puede usar cualquier libreria o herramienta para obtener el texto de su interés.



```python
#Importamos las librerias necesarias:
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import requests
from lxml import html
from bs4 import BeautifulSoup
```

```python
#Definimos la URL semilla:
url ='https://psicologiaymente.com/cultura/mejores-poemas-cortos'

#Realizamos la petición y parseamos el árbol HTML con Beautiful Soup:
respuesta=requests.get(url)
soup= BeautifulSoup(respuesta.text, features="lxml")
contenedor=soup.find(class_='prose md:prose-md xl:prose-xl content-other distance-1-5')
parrafo=contenedor.find('p').text
```

```python
#Creamos la nube de palabras:
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(parrafo)
```

```python
#Creamos el gráfico:
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```
